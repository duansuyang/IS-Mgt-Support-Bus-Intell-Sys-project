```python
import pandas as pd

# 按国家分组，并计算每个国家的平均收入和人数
country_stats = df1.groupby('Country').agg({'average_yearly_earnings': ['mean', 'count']})

# 重置列名
country_stats.columns = ['Average Income', 'Number of People']

# 创建新的 DataFrame 存储结果
result_df = pd.DataFrame({'Country': country_stats.index,
                          'Average Income': country_stats['Average Income'],
                          'Number of People': country_stats['Number of People']})

print(result_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 选择人数排名前10的国家，并按人均收入降序排序
top_10_countries = result_df.nlargest(10, 'Number of People').sort_values('Average Income', ascending=False)

# 设置画布大小和边距
fig, ax = plt.subplots(figsize=(12, 8))
plt.subplots_adjust(left=0.2, right=0.8, top=0.9, bottom=0.2)

# 绘制条形图
ax.bar(top_10_countries['Country'], top_10_countries['Average Income'])
ax.set_xlabel('Country')
ax.set_ylabel('Average Income')
ax.set_title('Countries by Average Income in Top 10 Countries by Number of Youtubers')
plt.xticks(rotation=45)

plt.savefig('国家中数量前10的人均收入排序.png', format='png')
print(top_10_countries["Country"])
```

```python
from wordcloud import WordCloud

df3 = df1.copy()

# Define remove & replace logic as replacing any non-alphanumeric characters
pattern = r'[^a-zA-Z0-9\s.,!?&\'-]'

# Replace characters 'Youtuber' & 'Title' with an empty string
df3['Youtuber'] = df3['Youtuber'].str.replace(pattern, '')
# Remove any trailing/ leading whitespace from values
df3['Youtuber'] = df3['Youtuber'].str.strip()
# Create filtered views of df to validate remove & replace
filter_youtuber_rows = df['Youtuber'].str.contains(pattern, regex = True)
# Validate results of string remove & replace
filter_youtuber_rows_results = filter_youtuber_rows[filter_youtuber_rows == True]


# Determine the 75th percentile of video views
percentile_75th = df3['video views'].quantile(0.75)
print(percentile_75th)
# Filter channels with views above the 75th percentile
high_view_channels = df3[df3['video views'] > percentile_75th]

# Combine the names of high-view channels into a single string
text = ' '.join(high_view_channels['Youtuber'])
print(text)
# Create the word cloud
wordcloud = WordCloud(background_color = 'white', width = 800, height = 400).generate(text)

# Plot the word cloud
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.figsize = (40,30)
plt.axis('off')
plt.title('Word Cloud of High-View YouTube Channel Descriptions', weight = 'bold')
plt.savefig("wordCloud.svg", format='svg')
```
